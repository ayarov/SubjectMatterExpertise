```python
import os
import numpy as np
import pandas as pd
```

```python
data_dir = r'D:\data\sme'
thesis_data_dir = os.path.join(data_dir, 'thesis')
train_file_path = os.path.join(thesis_data_dir, 'train_data.csv')

edit_types = ['edit_type_a', 
            'edit_type_b', 
            'edit_type_c', 
            'edit_type_d', 
            'edit_type_e', 
            'edit_type_f', 
            'edit_type_g', 
            'edit_type_h', 
            'edit_type_i', 
            'edit_type_j', 
            'edit_type_k', 
            'edit_type_l', 
            'edit_type_m']
```

```python
train_df = pd.read_csv(train_file_path)
```

```python
user_edits_df = pd.read_csv(os.path.join(data_dir, 'user_edits.csv'))
user_edits_df.head()
```

```python
edit_types_df = pd.read_hdf(os.path.join(data_dir, 'edit_types_2018.h5'))
edit_types_df.head()
```

```python
revisions_df = pd.read_csv(os.path.join(data_dir, 'revisions.csv'))
revisions_df.head()
```

```python
merged_df = revisions_df.merge(edit_types_df, on='revision_id', how='left')
print('Revisions shape: {}'.format(revisions_df.shape))
print('Edit types shape: {}'.format(edit_types_df.shape))
print('Merged shape: {}'.format(merged_df.shape))
```

```python
user_edits_map = {}
for i, row in user_edits_df.iterrows():
    user_name = row['user_name']
    total_edits = float(row['total_edits'])
    user_edits_map[user_name] = total_edits
```

```python
user_edits_map['Daniel Vaulot']
```

```python
counter = 0
data = []
for (page_id, user_name), group in merged_df.groupby(by=['page_id', 'user_name']):
    total_edits = user_edits_map[user_name]
    
    group = group.dropna(subset=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm'])
    
    data.append([page_id, 
                 user_name, 
                 group['a'].sum()/(total_edits),
                 group['b'].sum()/(total_edits),
                 group['c'].sum()/(total_edits),
                 group['d'].sum()/(total_edits),
                 group['e'].sum()/(total_edits), 
                 group['f'].sum()/(total_edits), 
                 group['g'].sum()/(total_edits), 
                 group['h'].sum()/(total_edits), 
                 group['i'].sum()/(total_edits), 
                 group['k'].sum()/(total_edits), 
                 group['k'].sum()/(total_edits), 
                 group['l'].sum()/(total_edits), 
                 group['m'].sum()/(total_edits)])
    
    if counter % 50000 == 0 and counter > 0:
        print(counter)
    counter += 1
user_edit_types_df = pd.DataFrame(data=data, columns=['page_id', 'user_name'] + edit_types)
```

```python
# print(user_edit_types_df.shape)
user_edit_types_df[user_edit_types_df.user_name == 'Daniel Vaulot'].head(100)
```

```python
user_edit_types_df.rename(columns={'user_name': 'user'}, inplace=True)
assert 'user' in user_edit_types_df.columns
```

```python
print(train_df.shape)
print(user_edit_types_df.shape)
```

```python
# train_no_edit_types_df = train_df.drop(labels=edit_types, axis=1)
merged_df = train_df.merge(user_edit_types_df, on=['page_id', 'user'], how='left')
merged_df.shape
```

```python
merged_df[merged_df.user == 'Daniel Vaulot'][['edit_type_i_x', 'edit_type_i_y']]
```

```python
edit_types_x = [etype + '_x' for etype in edit_types]
# print(edit_types_x)
edit_types_y = [etype + '_y' for etype in edit_types]
# print(edit_types_y)

edit_types_data = []
for i, row in merged_df.iterrows():
    page_id = row['page_id']
    user_name = row['user']
    index_data = [page_id, user_name]
    row = row[edit_types_y]
    if np.sum(row) == 0:
        row = merged_df[(merged_df.page_id == page_id) & (merged_df.user == user_name)][edit_types_x].iloc[0]  
    edit_types_data.append(index_data + row.get_values().tolist())
#     print(row.get_values().tolist()[0])
    complemented_user_edits_type_df = pd.DataFrame(data=edit_types_data, 
                                                   columns=['page_id', 'user'] + edit_types)
```

```python
complemented_user_edits_type_df[complemented_user_edits_type_df.user == 'Daniel Vaulot'].head()
# complemented_user_edits_type_df.dtypes
```

```python
complemented_user_edits_type_df.shape
```

```python
def calc_unclassified(df):
    unclassified = []
    for i, row in df[edit_types].iterrows():
        if np.sum(row) == 0:
            unclassified.append(i)    
    return unclassified
```

```python
total_unclassified = calc_unclassified(complemented_user_edits_type_df)
print('Total missing values: {} of {}'.format(len(total_unclassified), len(complemented_user_edits_type_df)))
```

```python
train_unclassified = calc_unclassified(train_df)
print('Total missing values: {} of {}'.format(len(train_unclassified), len(train_df)))

new_train_unclassified = calc_unclassified(merged_df)
print('Total missing values: {} of {}'.format(len(new_train_unclassified), len(merged_df)))
```
