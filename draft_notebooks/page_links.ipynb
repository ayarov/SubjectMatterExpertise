```python
import pymongo
import sqlparse
import pandas as pd
from wikipedia.mongo_tools import df_to_mongo
```

```python
def parse_sql_query(sql):
    data = []
    if sql.startswith('INSERT INTO `pagelinks` VALUES'):
        parsed = sqlparse.parse(sql=sql)

        for token in parsed[0].tokens:
            if isinstance(token, sqlparse.sql.Values):
                for value_token in token.tokens:
                    if isinstance(value_token, sqlparse.sql.Parenthesis):
                        segments = value_token.value.split(',')
                        if len(segments) == 1:
                            print(segments)
                        else:
                            try:
                                pl_from = int(segments[0].replace('(', ''))
                                pl_namespace = int(segments[1])
                                pl_from_namespace = int(segments[len(segments)-1].replace(')', ''))
                                
                                title_segments = segments[2:len(segments)-1]
                                
                                pl_title = str.join('', title_segments)
                                data.append([pl_from, pl_namespace, pl_title, pl_from_namespace])
                            except ValueError as ex:
                                print(f'Error: {ex}\tValue: {segments}')
    return data
```

```python
mongo_host = 'localhost'
mongo_port = 27017
mongo_db = 'enwiki'
mongo_collection = 'page_links'
lines = []
with open(r'../../../Downloads/enwiki-latest-pagelinks.sql', encoding='ISO-8859-1') as f:

    with pymongo.MongoClient(host=mongo_host, port=mongo_port) as client:
        db = client.get_database(mongo_db)
        collection = db.get_collection(mongo_collection)
        for sql in f:
            data = parse_sql_query(sql=sql)
            df = pd.DataFrame(data=data, columns=['pl_from', 'pl_namespace', 'pl_title', 'pl_from_namespace'])
            df_to_mongo(df=df, collection=collection)
```
