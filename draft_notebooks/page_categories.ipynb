```python
import pymongo
import sqlparse
import pandas as pd
from wikipedia.mongo_tools import df_to_mongo
```

```python
def parse_sql_query(sql):
    data = []
    if sql.startswith('INSERT INTO `categorylinks` VALUES'):
        parsed = sqlparse.parse(sql=sql)

        for token in parsed[0].tokens:
            if isinstance(token, sqlparse.sql.Values):
                for value_token in token.tokens:
                    if isinstance(value_token, sqlparse.sql.Parenthesis):
                        segments = value_token.value.split(',')
                        if len(segments) == 1:
                            print(segments)
                        else:
                            page_id = segments[0]
                            category = segments[1]
                            data.append([int(page_id.replace('(', '')), str(category[1:len(category)-1])])
    return data
```

```python
mongo_host = 'localhost'
mongo_port = 27017
mongo_db = 'enwiki'
mongo_collection = 'page_categories'
lines = []
with open(r'../../../Downloads/enwiki-latest-categorylinks.sql', encoding='ISO-8859-1') as f:

    with pymongo.MongoClient(host=mongo_host, port=mongo_port) as client:
        db = client.get_database(mongo_db)
        collection = db.get_collection(mongo_collection)
        for sql in f:
            data = parse_sql_query(sql=sql)
            df = pd.DataFrame(data=data, columns=['page_id', 'category'])
            df_to_mongo(df=df, collection=collection)
```
