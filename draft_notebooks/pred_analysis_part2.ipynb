```python
test_df = test_df[['expert', 
                   'category', 
                   'page', 
                   'page_id', 
                   'user', 
                   'registered', 
                   'evaluation', 
                   'is_wikipedian', 
                   'is_admin', 
                   'has_user_page']]

test_df['article'] = test_df['page']
test_df['article_id'] = test_df['page_id']
test_df['is_wikipedian'] = test_df.apply(lambda row: False if pd.isnull(row['is_wikipedian']) or row['is_wikipedian'] == 0 else True, axis=1)
test_df['is_admin'] = test_df.apply(lambda row: False if pd.isnull(row['is_admin']) or row['is_admin'] == 0 else True, axis=1)
test_df['has_user_page'] = test_df.apply(lambda row: True if pd.isnull(row['has_user_page']) or row['has_user_page'] > 0 else False, axis=1)

test_df.drop(['page', 'page_id'], axis=1)
```

```python
df = test_df.merge(activity_scopes_df, how='left', on='article_id')
df[df.registered].head(100)
```

```python
rows = []
grouped = df.groupby('article_id')
for article_id, group in grouped:
    category = group['category'].iloc[0]
    article = group['article'].iloc[0]    
    maturity = group['activity_scope'].iloc[0]
    total_registered = len(group[group.registered])
    total_unregistered = len(group[~group.registered])
    total_unique = len(group['user'].unique())
    
    sorted_group = group.sort_values(by='expert', ascending=False)
    
    top_10 = sorted_group.head(10)
    top_10_registered = len(top_10[sorted_group.registered==True])
    top_10_unregistered = len(top_10[sorted_group.registered==False])
    
    top5_registered = sorted_group[sorted_group.registered].head(5)
    top5_80_registered = sorted_group[(sorted_group.registered) & (sorted_group.expert > 0.8)].head(5)
    
    top5_expert = len(top5_registered[top5_registered['evaluation']=='expert'])
    top5_no_page = len(top5_registered[top5_registered['evaluation']=='no_page'])
    top5_not_expert = len(top5_registered[top5_registered['evaluation']=='not_expert'])
    top5_practitioner = len(top5_registered[top5_registered['evaluation']=='practitioner'])
    top5_scholar = len(top5_registered[top5_registered['evaluation']=='scholar'])
    top5_unknown = len(top5_registered[top5_registered['evaluation']=='unknown'])
    top5_wikipedian = len(top5_registered[top5_registered['is_wikipedian']])
    top5_administrator = len(top5_registered[top5_registered['is_admin']])
    top5_has_user_page = len(top5_registered[top5_registered['has_user_page']])
    top5_nan = len(top5_registered[top5_registered['evaluation'].isnull()])
    
    top5_80_expert = len(top5_80_registered[top5_80_registered['evaluation']=='expert'])
    top5_80_no_page = len(top5_80_registered[top5_80_registered['evaluation']=='no_page'])
    top5_80_not_expert = len(top5_80_registered[top5_80_registered['evaluation']=='not_expert'])
    top5_80_practitioner = len(top5_80_registered[top5_80_registered['evaluation']=='practitioner'])
    top5_80_scholar = len(top5_80_registered[top5_80_registered['evaluation']=='scholar'])
    top5_80_unknown = len(top5_80_registered[top5_80_registered['evaluation']=='unknown'])
    top5_80_wikipedian = len(top5_80_registered[top5_80_registered['is_wikipedian']])
    top5_80_administrator = len(top5_80_registered[top5_80_registered['is_admin']])
    top5_80_has_user_page = len(top5_80_registered[top5_80_registered['has_user_page']])
    top5_80_nan = len(top5_80_registered[top5_80_registered['evaluation'].isnull()])
        
    rows.append([category, 
                 article, 
                 article_id, 
                 maturity, 
                 total_unique, 
                 total_registered, 
                 total_unregistered, 
                 top_10_registered, 
                 top_10_unregistered,
                 top5_expert,
                 top5_not_expert,
                 top5_practitioner,
                 top5_scholar,
                 top5_unknown,
                 top5_wikipedian,
                 top5_administrator,
                 top5_has_user_page,
                 top5_nan,
                 top5_80_expert,
                 top5_80_not_expert,
                 top5_80_practitioner,
                 top5_80_scholar,
                 top5_80_unknown,
                 top5_80_wikipedian,
                 top5_80_administrator,
                 top5_80_has_user_page,
                 top5_80_nan])

new_df = pd.DataFrame(rows, columns=['category', 
                                     'article',
                                     'article_id', 
                                     'maturity', 
                                     '#distinct_users',
                                     '#registered_users',
                                     '#unregistered_users',
                                     '#top10_registered_users',
                                     '#top10_unregistered_users',
                                     'top5_expert',
                                     'top5_not_expert',
                                     'top5_practitioner',
                                     'top5_scholar',
                                     'top5_unknown',
                                     'top5_wikipedian',
                                     'top5_administrator',
                                     'top5_has_user_page',
                                     'top5_nan',
                                     'top5_80_expert',
                                     'top5_80_not_expert',
                                     'top5_80_practitioner',
                                     'top5_80_scholar',
                                     'top5_80_unknown',
                                     'top5_80_wikipedian',
                                     'top5_80_administrator',
                                     'top5_80_has_user_page',
                                     'top5_80_nan'])
new_df = new_df.sort_values(by='category', ascending=True)
new_df.head()
```

```python
def average_by_column(column):
    avg_rows = []
    avg_cols = [col for col in new_df.columns if col not in ['category', 'article', 'article_id', 'maturity']]
    groups_by_column = new_df.groupby(column)
    for group_column, group in groups_by_column:
        avg_row = [group_column]
        for col in avg_cols:
            avg_row.append(np.round(np.average(group[col]), 2))
        avg_rows.append(avg_row)
    return pd.DataFrame(avg_rows, columns=[column] + avg_cols) 
```

```python
by_category_df = average_by_column('category')
by_category_df.to_excel(r'd:/temp/avg_by_category_v2.xlsx', index=False)
```

```python
by_maturity_df = average_by_column('maturity')
by_maturity_df.to_excel(r'd:/temp/avg_by_maturity_v2.xlsx', index=False)
```
