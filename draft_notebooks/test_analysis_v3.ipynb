```python
test_df = test_df[test_df.page_edits > 2]
total_edits = np.round(np.sum(test_df['page_edits']), 2)
total_population = len(test_df)

stats_cols = ['Threshold', 
              'Total Experts', 
              'Total Experts Edits', 
              '% Experts', 
              '% Experts Edits',
              'Total Editors',
              'Total Edits']
stats_rows = []
for i in range(70, 91, 1):
    threshold = i / 100
    total_experts = len(test_df[test_df.expert > threshold])
    total_experts_edits = np.sum(test_df[test_df.expert > threshold]['page_edits'])
    stats_rows.append([threshold, total_experts, 
                       total_experts_edits, 
                       total_experts/total_population, 
                       total_experts_edits/total_edits,
                       total_population,
                       total_edits])
    
stats_df = pd.DataFrame(stats_rows, columns=stats_cols)
stats_df.to_excel(r'd:/temp/sensitivity_analysis_v3a.xlsx', index=False)
```

```python
test_df = test_df[['expert', 
                   'category', 
                   'page', 
                   'page_id', 
                   'user', 
                   'registered', 
                   'evaluation', 
                   'is_wikipedian', 
                   'is_admin', 
                   'is_practitioner',
                   'is_scholar',
                   'has_user_page']]

test_df['article'] = test_df['page']
test_df['article_id'] = test_df['page_id']
test_df['is_wikipedian'] = test_df.apply(lambda row: False if pd.isnull(row['is_wikipedian']) or row['is_wikipedian'] == 0 else True, axis=1)
test_df['is_admin'] = test_df.apply(lambda row: False if pd.isnull(row['is_admin']) or row['is_admin'] == 0 else True, axis=1)
test_df['is_practitioner'] = test_df.apply(lambda row: False if pd.isnull(row['is_practitioner']) or row['is_practitioner'] == False else True, axis=1)
test_df['is_scholar'] = test_df.apply(lambda row: False if pd.isnull(row['is_scholar']) or row['is_scholar'] == False else True, axis=1)
test_df['has_user_page'] = test_df.apply(lambda row: True if pd.isnull(row['has_user_page']) or row['has_user_page'] > 0 else False, axis=1)

test_df = test_df.drop(['page', 'page_id'], axis=1)
```

```python
def average_by_column(column):
    avg_rows = []
    avg_cols = [col for col in new_df.columns if col not in ['category', 'article', 'article_id', 'maturity']]
    groups_by_column = new_df.groupby(column)
    for group_column, group in groups_by_column:
        avg_row = [group_column]
        for col in avg_cols:
            avg_row.append(np.round(np.average(group[col]), 2))
        avg_rows.append(avg_row)
    return pd.DataFrame(avg_rows, columns=[column] + avg_cols) 
```

```python
by_category_df = average_by_column('category')
by_category_df.to_excel(r'd:/temp/avg_by_category_v3.xlsx', index=False)
```

```python
by_maturity_df = average_by_column('maturity')
by_maturity_df.to_excel(r'd:/temp/avg_by_maturity_v3.xlsx', index=False)
```
