```python
%autosave 10000
# %matplotlib inline

import os
import numpy as np
import pandas as pd
# import seaborn as sns
# import matplotlib.pyplot as plt

# plt.style.use('seaborn-colorblind')
# sns.palplot(sns.color_palette("GnBu_d"))

pd.options.display.max_rows = 20
```

```python
train_data = pd.read_csv(r'../../../data/sme/thesis/train_data.csv', encoding='utf-8')
test_data = pd.read_csv(r'../../../data/sme/thesis/test_data.csv', encoding='utf-8')
test_data_385 = pd.read_pickle(path=r'../../../data/sme/thesis/test_data_385_articles.bz2', compression='bz2')
```

```python
edit_type_columns = {'a': 'edit_type_a', 
                     'b': 'edit_type_b', 
                     'c': 'edit_type_c', 
                     'd': 'edit_type_d', 
                     'e': 'edit_type_e', 
                     'f': 'edit_type_f', 
                     'g': 'edit_type_g', 
                     'h': 'edit_type_h', 
                     'i': 'edit_type_i', 
                     'j': 'edit_type_j', 
                     'k': 'edit_type_k', 
                     'l': 'edit_type_l', 
                     'm': 'edit_type_m'}
```

```python
revisions = pd.read_csv(r'../../../data/sme/revisions.csv')
revisions.set_index(keys='revision_id', inplace=True)
print(revisions.shape)
revisions.head()
```

```python
edit_types_2012 = pd.read_hdf(r'../../../data/sme/edit_types_2012.h5')
edit_types_2012.rename(columns=edit_type_columns, inplace=True)
edit_types_2012.set_index(keys='revision_id', inplace=True)
print(edit_types_2012.shape)
edit_types_2012.head()
```

```python
edit_types_2018 = pd.read_hdf(r'../../../data/sme/edit_types_2018.h5')
edit_types_2018.rename(columns=edit_type_columns, inplace=True)
edit_types_2018.set_index(keys='revision_id', inplace=True)
print(edit_types_2018.shape)
edit_types_2018.head()
```

```python
rev_edit_types_2012 = revisions.join(edit_types_2012, how='left')
print(rev_edit_types_2012.shape)
rev_edit_types_2012.head()
```

```python
rev_edit_types_2018 = revisions.join(edit_types_2018, how='left')
print(rev_edit_types_2018.shape)
rev_edit_types_2018.head()
```

```python
edit_type_cols = list(edit_type_columns.values())
rev_cols = rev_edit_types_2018.drop(columns=edit_type_columns.values()).columns.tolist()
print(rev_cols)
final_cols = [rev_edit_types_2018.index.name] + rev_cols + edit_type_cols
print(final_cols)
```

```python
suffix_2018 = '_x'
suffix_2012 = '_y'
rev_edit_types = rev_edit_types_2018.join(rev_edit_types_2012[edit_type_cols], 
                                          how='left', 
                                          lsuffix=suffix_2018, 
                                          rsuffix=suffix_2012)
print(rev_edit_types.shape)
rev_edit_types.head()
```

```python
for et_col in edit_type_cols:
    print('Processing: {}'.format(et_col))
    rev_edit_types[et_col] = rev_edit_types.apply(lambda row: row[et_col+suffix_2012] 
                                                  if pd.isnull(row[et_col+suffix_2018]) 
                                                  else row[et_col+suffix_2018], axis=1)
```

```python
drop_list_2018 = [et_col + suffix_2018 for et_col in edit_type_cols]
drop_list_2012 = [et_col + suffix_2012 for et_col in edit_type_cols]
rev_edit_types.drop(columns=drop_list_2018 + drop_list_2012, inplace=True)
rev_edit_types.head()
```

```python
rev_edit_types_2012.reset_index(inplace=True)
rev_edit_types_2012.head()
```

```python
rev_edit_types_2018.reset_index(inplace=True)
rev_edit_types_2018.head()
```

```python
rev_edit_types_2012.to_csv(r'../../../data/sme/thesis/revisions_with_edit_types_2012.csv', index=False)
```

```python
rev_edit_types_2018.to_csv(r'../../../data/sme/thesis/revisions_with_edit_types_2018.csv', index=False)
```

```python
rev_edit_types.to_csv(r'../../../data/sme/thesis/revisions_with_edit_types_final.csv', index=False)
```

```python
counter = 0
data = []
for (page_id, user_name), group in df.groupby(by=['page_id', 'user_name']):
#     total_edits = user_edits_map[user_name]
    
    row = [page_id, user_name] + [np.sum(group[et_col])/len(group) for et_col in edit_type_cols]
    data.append(row)

    if counter % 50000 == 0 and counter > 0:
        print(counter)
    counter += 1
    
user_edit_types_df = pd.DataFrame(data=data, columns=['page_id', 'user_name'] + edit_type_cols)
user_edit_types_df.rename(columns={'user_name': 'user'}, inplace=True)
```

```python
train_data_no_et = train_data.drop(columns=edit_type_cols)
test_data_no_et = test_data.drop(columns=edit_type_cols)
test_data_385_no_et = test_data_385.drop(columns=edit_type_cols)

print(train_data_no_et.shape)
print(test_data_no_et.shape)
print(test_data_385_no_et.shape)
```

```python
new_train_data = train_data_no_et.merge(user_edit_types_df, how='left', on=['page_id', 'user'])
print(new_train_data.shape)
```

```python
new_test_data = test_data_no_et.merge(user_edit_types_df, how='left', on=['page_id', 'user'])
print(new_test_data.shape)
```
